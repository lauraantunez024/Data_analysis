## Monthly Mean Graphs

This subset of data is the first six months of the dataframe showing the monthly means, the x-axis is the year, and the y axis is the temperature.

```python
months = monthly_means[['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']]

for x in months:
    plt.plot(monthly_means.Year,months[x])
    plt.title(x)
    plt.xlabel('Year')
    plt.ylabel('Temperaures (in degrees)')
    plt.show()
```

### Conclusions
 * Interesting findings: Graphics show that around 1940, temperatures began to rise exponentially in all the months shown, and then fell just as quickly a few years later. Some months are getting hotter than others in more recent years, such as February and March. 
 * What data shows: Monthly data also further proves that temperatures are rising, and the winter months are the ones that are most effected.
